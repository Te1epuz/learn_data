```python
import pandas as pd
```

```python
pd.options.display.float_format = '{:.2f}'.format # меняем дефолт вид чисел с плавающей запятой
```

```python
export_2015 = pd.read_excel('russia2015.xlsx', sheet_name='Partner')
export_2016 = pd.read_excel('russia2016.xlsx', sheet_name='Partner')
export_2017 = pd.read_excel('russia2017.xlsx', sheet_name='Partner')
```

```python
export_2015.head()
```

```python
export = pd.concat([export_2015, export_2016, export_2017])
```

```python
export.sample(5) # выводит случайные строки
```

```python
exclude = ['Europe & Central Asia', 'East Asia & Pacific', ' World',
    'Unspecified', 'Middle East & North Africa', 'North America', 'Latin America & Caribbean', 
    'Other Asia, nes', 'South Asia']
export_filtered = export[-export['Partner Name'].isin(exclude)]
```

```python
export_pivot = export_filtered.pivot_table(index=['Year', 'Partner Name'], values='Export (US$ Thousand)', aggfunc='sum')\
    .sort_values(['Year', 'Export (US$ Thousand)'], ascending=[True, False])
```

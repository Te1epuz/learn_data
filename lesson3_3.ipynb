```python
import json
```

```python
metro_stations = []

with open('metro.json', 'r', encoding='cp1251') as f:
    datastore = json.load(f)

for rows in datastore:
    station = {}
    station['станция'] = rows['NameOfStation']
    station['долгота'] = rows['Longitude_WGS84']
    station['широта'] = rows['Latitude_WGS84']
    #station['координаты'] = rows['geoData']['coordinates']
    
    if not station in metro_stations:
        metro_stations.append(station)

print(datastore[0])

metro_stations
```
